```python
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
```

```python
# stardf = pd.read_csv("allfaintstars.csv")
# galaxydf = pd.read_csv("allfaintgalaxies.csv")
# df = pd.concat([stardf,galaxydf]).reset_index(drop=True)

df = pd.read_csv("faint22test.csv")
```

```python
df = df.replace(-9999, np.NaN).dropna().reset_index(drop=True)
print(f"Total of {len(df)} objects")
df["class"].value_counts()
```

```python
filters = "ugriz"
links_arr = []
fname_dict = {}
```

```python
for row in tqdm(df.itertuples(), total=len(df)):
    objid = row[1]
    obj_class = row[3]
    run = row[4]
    run6 = "0"*(6 - len(str(run)))+str(run)
    rerun = row[5]
    camcol = row[6]
    field = row[7]
    field4 = "0"*(4 - len(str(field)))+str(field)
    ra = row[8]
    dec = row[9]
    inner_obj_dict = {"ra":ra, "dec":dec,"field":field, "field4":field4, "camcol":camcol, "rerun":rerun, "run6":run6, "run":run,"obj_class":obj_class}

    for filter_letter in filters:
        fits_url = f"https://data.sdss.org/sas/dr16/eboss/photoObj/frames/{rerun}/{run}/{camcol}/frame-{filter_letter}-{run6}-{camcol}-{field4}.fits.bz2"
        fname = f"frame-{filter_letter}-{run6}-{camcol}-{field4}.fits.bz2"
        links_arr.append(fits_url)
        inner_obj_dict[f"url_{filter_letter}"]=fits_url
        inner_obj_dict[f"fname_{filter_letter}"]=fname
    fname_dict[objid] = inner_obj_dict
```

```python
import os
import pickle
os.chdir("allcondsdata")
```

```python
# Make Sure To Link Google Drive
with open(f"sdss_dictionary_pickle_allconds", "wb") as fp:
    pickle.dump(fname_dict, fp)
#with open(f"sdss_fitslinks_v{version_num}.txt", "w") as outfile:
#    outfile.write("\n".join(links_arr))

df.to_csv(f"SDSS_Query_allconds_DF.csv",index=False)
```
